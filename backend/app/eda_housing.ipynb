```python
# Target variable
y = df_encoded['median_house_value'].values

# Features (all columns except target)
X = df_encoded.drop('median_house_value', axis=1).values

# Or get feature names for later
feature_names = df_encoded.drop('median_house_value', axis=1).columns.tolist()
```
